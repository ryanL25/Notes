# Robust Optimization


Original Optimization Problem: 





\begin{align*}
\max_{x,y}: \quad & 12x + 9y \\
\text{s.t}: \quad & x \le 1000 \\
		  & y\le 1500 \\
		  & x + y \le 1750\\
		  & 4x + 2y \le 4800 \\
		  & x,y \ge 0
\end{align*}



Robust Optimization Problem: 

Copper for logic chips now between 3.5g and 4.5g, memory chips now between 1.8g to 2.2g. 


\begin{align*}
\max_{x,y}: \quad & 12x + 9y \\
\text{s.t}: \quad & x \le 1000 \\
		  & y\le 1500 \\
		  & x + y \le 1750\\
		  & \max_{\delta_x \in [-.5, .5], \delta_y \in [-.2, .2]} (4 + \delta_X) x + (2 + \delta_y)y  \le 4800 \\
		  & x,y \ge 0
\end{align*}

Linear Reformulation


\begin{align*}
\max_{x,y}: \quad & 12x + 9y \\
\text{s.t}: \quad & x \le 1000 \\
		  & y\le 1500 \\
		  & x + y \le 1750\\
		  & 4.5x + 2.2y  \le 4800 \\
		  & x,y \ge 0
\end{align*}

```python
import pyomo.environ as pyo


SOLVER = pyo.SolverFactory('appsi_highs')
assert SOLVER.available(), "HiGHS (appsi_highs) solver is not available."
print("Solver ready:", SOLVER)
```

```python
model = pyo.ConcreteModel()

model.x = pyo.Var(domain=pyo.NonNegativeReals)
model.y = pyo.Var(domain=pyo.NonNegativeReals)

model.obj = pyo.Objective(expr=3*model.x + 2*model.y, sense=pyo.maximize)

model.con1 = pyo.Constraint(expr= model.x <= 1000)
model.con2 = pyo.Constraint(expr= model.y <= 1500)
model.con3 = pyo.Constraint(expr = model.x + model.y <= 1750) 
model.con4 = pyo.Constraint(expr = 4.5*model.x + 2.2*model.y <= 4800)

results = SOLVER.solve(model)
```

```python
print(f"Optimal value: {model.obj()}")
print(f"x: {model.x()}, y: {model.y()}")
```



\begin{align*}
\max_{x,y}: \quad & 12x + 9y \\
\text{s.t}: \quad & x \le 1000 \\
		  & y\le 1500 \\
		  & x + y \le 1750\\
		  & 4.5x + 2.2y  \le 4800 \\
		  & x,y \in \{0,1\}
\end{align*}

```python
model = pyo.ConcreteModel()

model.x = pyo.Var(domain=pyo.NonNegativeIntegers)
model.y = pyo.Var(domain=pyo.NonNegativeIntegers)

model.obj = pyo.Objective(expr=3*model.x + 2*model.y, sense=pyo.maximize)

model.con1 = pyo.Constraint(expr= model.x <= 1000)
model.con2 = pyo.Constraint(expr= model.y <= 1500)
model.con3 = pyo.Constraint(expr = model.x + model.y <= 1750) 
model.con4 = pyo.Constraint(expr = 4.5*model.x + 2.2*model.y <= 4800)

results = SOLVER.solve(model)
```

```python
print(f"Optimal value: {model.obj()}")
print(f"x: {model.x()}, y: {model.y()}")
# floating point inaccuracy, rounding to nearest integer
```
