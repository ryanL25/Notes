We want to max 

\begin{align*}
\max \quad &x-3e^x \\
\text{s.t.} \quad &-3\le x\le 5
\end{align*}

With $\epsilon \leq 10^{-5}$

```python
import math

def func(x):
    return x - 3 * math.exp(x)
a = -3
b = 5
error = b - a

phi = 1 / ((1 + math.sqrt(5)) / 2)

# get inital values
x1 = a + (1-phi) * (b-a)
x2 = a + phi * (b-a)

#iterate until error is satisfied 
while error > 10e-5:
    if func(x1) < func(x2):
        # cut off right side of interval 
        a = x1 
        x1 = x2 
        x2 = a + phi * (b-a)
    else:
        #cut off left side of interval 
        b = x2 
        x2 = x1 
        x1 = a + (1-phi) * (b-a)
    
    error = b-a 

optimal_x_gss = (a+b)/2
print("The max value is at x =", optimal_x_gss)
    
```

```python
import cvxpy as cp

x = cp.Variable()
func = x - 3 * cp.exp(x) 

problem = cp.Problem(cp.Maximize(func), [x >= -3, x <= 5])
result = problem.solve()
optimal_x_cvx = x.value 

print("The max value is at x =", optimal_x_cvx)
```

```python
print(f"Different in solutions is {abs(optimal_x_gss - optimal_x_cvx)}")
```
