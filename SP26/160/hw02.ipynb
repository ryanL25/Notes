```python
import numpy as np
```

```python
points = np.array([
    [0, 0],
    [0, 4],
    [2, 0],
    [4, 5],
    [-10,10]
])
```

```python
x = points[:, 0]
y = points[:, 1]
```

```python
x4 = sum(x ** 4)
y4 = sum(y ** 4)
x2 = sum(x ** 2)
y2 = sum(y ** 2)
x2y2 = sum(x ** 2 * y ** 2)

print(f"x4: {x4}")
print(f"y4: {y4}")
print(f"x2: {x2}")
print(f"y2: {y2}")
print(f"x2y2: {x2y2}")
```
